```python
#Question 4
import numpy as np

# Demand data for silver and black mobile phones
silver_demand = [250, 350, 400, 150, 50, 200, 250, 100]
black_demand = [200, 200, 50, 210, 330, 180, 80, 250]

# Coefficient of correlation
correlation_coefficient = np.corrcoef(silver_demand, black_demand)[0, 1]

# Mean demand
silver_mean = np.mean(silver_demand)
black_mean = np.mean(black_demand)

# Standard deviation of demand
silver_std = np.std(silver_demand, ddof=1)
black_std = np.std(black_demand, ddof=1)
# Print the results
print("Silver Mobile Phones:")
print(f"Coefficient of Correlation: {correlation_coefficient:.2f}")
print(f"Mean Demand: {silver_mean:.2f}")
print(f"Standard Deviation: {silver_std:.2f}")
print()

print("Black Mobile Phones:")
print(f"Coefficient of Correlation: {correlation_coefficient:.2f}")
print(f"Mean Demand: {black_mean:.2f}")
print(f"Standard Deviation: {black_std:.2f}")


#b

import scipy.stats as stats

unit_revenue = 88  # Unit revenue for both colors
unit_procurement_price = 55  # Unit procurement price for both colors
spare_parts_price = 30  # Price for selling leftover inventory as spare parts


# Optimal order quantity for silver mobile phones
cr=(unit_revenue-unit_procurement_price)/(unit_revenue-spare_parts_price)

z = stats.norm.ppf(cr)
silver_optimal_order_quantity=silver_mean+z*silver_std  # Assuming desired service level of 90%

# Optimal order quantity for black mobile phones
black_optimal_order_quantity = stats.norm.ppf(cr, black_mean, black_std)  # Assuming desired service level of 90%

# Calculate the expected profit for silver mobile phones
from scipy.stats import norm
g_z=norm.pdf(z)-z*(1-norm.cdf(z))
silver_ELS=silver_std*g_z
silver_ES=silver_mean-silver_ELS
silver_ELO=silver_optimal_order_quantity-silver_ES
silver_EP=-unit_procurement_price*silver_optimal_order_quantity+unit_revenue*silver_ES+silver_ELO*spare_parts_price


black_ELS=black_std*g_z
black_ES=black_mean-black_ELS
black_ELO=black_optimal_order_quantity-black_ES
black_EP=-unit_procurement_price*black_optimal_order_quantity+unit_revenue*black_ES+black_ELO*spare_parts_price


# Print the results
print("Silver Mobile Phones:")
print(f"Optimal Order Quantity: {silver_optimal_order_quantity:.2f} units")
print(f"Expected Profit: {silver_EP:.2f} €")
print()

print("Black Mobile Phones:")
print(f"Optimal Order Quantity: {black_optimal_order_quantity:.2f} units")
print(f"Expected Profit: {black_EP:.2f} €")

print("Total Profit:",silver_EP+black_EP)


#c
demand = [x + y for x, y in zip(silver_demand,black_demand)]
mean = np.mean(demand)
std = np.std(demand, ddof=1)

price=88
common_order_cost=50
color_order_cost=5
salvage=29

underage_cost=price-common_order_cost-color_order_cost
overage_cost=common_order_cost-salvage
cr=underage_cost/(underage_cost+overage_cost)
z = stats.norm.ppf(cr)
order_quantity = stats.norm.ppf(cr, mean, std)  # Assuming desired service level of 90%

g_z=norm.pdf(z)-z*(1-norm.cdf(z))
ELS=std*g_z
ES=mean-ELS
ELO=order_quantity-ES
EP=underage_cost*ES+ELO*(salvage-common_order_cost)

```
